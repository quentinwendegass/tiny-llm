```python
import matplotlib.pyplot as plt
import torch

from src.helper.warmup_scheduler import WarmupCosineAnnealingScheduler
```

```python
warmup_steps = 2000
learning_rate = 6e-4
max_steps = 20000
```

```python
model = torch.nn.Linear(2, 2)
optimizer = torch.optim.Adam(model.parameters())
scheduler = WarmupCosineAnnealingScheduler(optimizer, max_steps, warmup_steps, learning_rate)

x = range(max_steps)
y = [scheduler.rate(i) for i in x]

plt.plot(x,y)
```
